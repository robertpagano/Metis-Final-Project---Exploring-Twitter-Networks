```python
def create_df(json_filename):
    # inputs a json file, converts to json object, then runs the data clean function
    # will need to loop through all json files, and then merge the dataframes
    with open(json_filename) as json_file:
        tweets = json.load(json_file)
    return deefify2(tweets)
    
```
