Below I am loading json files for each of my 391 json files. Once I am done with that:
 - munge all using deefify2
 - merge all files
 - remove any duplicate tweet IDs
 
Then that's my final dataframe with all data. From there I can select the data I want to run my graph

```python
def create_df(json_filename):
    # inputs a json file, converts to json object, then runs the data clean function
    # will need to loop through all json files, and then merge the dataframes
    with open(json_filename) as json_file:
        tweets = json.load(json_file)
    return deefify2(tweets)
    
```
