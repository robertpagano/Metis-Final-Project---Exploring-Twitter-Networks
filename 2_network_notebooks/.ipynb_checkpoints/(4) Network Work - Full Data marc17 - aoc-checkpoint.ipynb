```python
with open('/Users/robertpagano/metis_data/project_5/final_raw_df.pickle', 'rb') as f:
    df_aoc = pickle.load(f)
```

```python
df_aoc.head()
```

```python
df_aoc_names = df_aoc.filter(['screen_name','retweet_screen_name', 'rt_count', 'retweet_fav_count', 'retweet_followers_count', 'retweet_is_verified'], axis=1)
```

```python
print(df_aoc_names.shape)
df_aoc_names = df_aoc_names.dropna(axis=0, subset=['retweet_screen_name'])
print(df_aoc_names.shape)
```

Just a note - come back to the above if I need to add more information to my network data

Also - I'm just going to go with retweets first this time. Retweets more often imply someone is agreeing with what is being said, while a quote is often someone disagreeing

```python
df_aoc_names.sample(10)
```

```python
# First creating undirected:
G_aoc_und = nx.from_pandas_edgelist(df_aoc_names, source='screen_name', target='retweet_screen_name', edge_attr=True, create_using=None)

# Second creating directed:
G_aoc_dir = nx.from_pandas_edgelist(df_aoc_names, source='screen_name', target='retweet_screen_name', edge_attr=True, create_using=nx.DiGraph())
```

```python
print(nx.info(G_aoc_und))
print(nx.info(G_aoc_dir))
```

```python
# removing all self re-tweets
G_aoc_und.remove_edges_from(G_aoc_und.selfloop_edges())

```

```python
print(nx.info(G_aoc_und))
print(nx.info(G_aoc_dir))
```

That removed a few thousand - also, I am going to stick with undirected, because the next step undirects it anway

```python
## The biggest component is way bigger than every other one, so we can just focus on that for the graph
G_aoc_und_subgraphs = sorted(nx.connected_component_subgraphs(G_aoc_und.to_undirected()), key=len, reverse=True)
print('Size of largest connected components:', [len(g) for g in G_aoc_und_subgraphs[:20]])
```

```python
G_aoc_dir_subgraphs = sorted(nx.connected_component_subgraphs(G_aoc_dir.to_undirected()), key=len, reverse=True)
print('Size of largest connected components:', [len(g) for g in G_aoc_dir_subgraphs[:20]])
```

```python
G_aoc_undir_final = G_aoc_und_subgraphs[0]
```

```python
# Determine which community each user is in
G_aoc_partition = community.best_partition(G_aoc_und_subgraphs[0].to_undirected())
```

```python
G_aoc_partition
```

```python
max(G_aoc_partition.values())
```

```python
# Find strongly connected components
G_conn_comp = sorted(nx.strongly_connected_components(G_aoc_dir),key=len, reverse=True)
print('Size of largest strongly connected components:', [len(g) for g in G_conn_comp[:20]])
```

```python
# Create graph of largest strongly connected component, this is just for the directed graph
S = nx.subgraph(G_aoc_dir, G_conn_comp[0])
```

```python
#Calculating centrality measures on the main graph object. All saved as a dict object
pr = nx.pagerank(S)
bt = nx.betweenness_centrality(S)
cl = nx.closeness_centrality(S)
```

```python
## pickling each for safe keeping since it takes so long to run (roughly 20-30 minutes)

with open('/Users/robertpagano/src/metis_project_kojak/network_files/aoc/aoc_pagerank.pickle', 'wb') as to_write:
    pickle.dump(pr, to_write)
    
with open('/Users/robertpagano/src/metis_project_kojak/network_files/aoc/aoc_betweenness.pickle', 'wb') as to_write:
    pickle.dump(bt, to_write)
    
with open('/Users/robertpagano/src/metis_project_kojak/network_files/aoc/aoc_closeness.pickle', 'wb') as to_write:
    pickle.dump(cl, to_write)
    
# with open('/Users/robertpagano/src/metis_project_kojak/network_files/aoc/aoc_maingraph_dir.pickle', 'wb') as to_write:
#     pickle.dump(S, to_write)
    
with open('/Users/robertpagano/src/metis_project_kojak/network_files/aoc/aoc_maingraph_undir.pickle', 'wb') as to_write:
    pickle.dump(G_aoc_undir_final, to_write)
    
with open('/Users/robertpagano/src/metis_project_kojak/network_files/aoc/aoc_maingraph_partition.pickle', 'wb') as to_write:
    pickle.dump(G_aoc_partition, to_write)
    
```
