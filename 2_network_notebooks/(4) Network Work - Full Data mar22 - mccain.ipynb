```python
df_mccain_names = df_mccain.filter(['screen_name', 'retweet_screen_name', 'reply_to_screen_name', 'quoted_screen_name', 'tweet_id', 'rt_count', 'retweet_fav_count', 'quoted_fav_count', 'retweet_followers_count', 'retweet_is_verified', 'quoted_is_verified'], axis=1)
```

```python
df_mccain_names.head()
```

```python
df_mccain_names.reply_to_screen_name.fillna(value=pd.np.nan, inplace=True)
```

```python
df_mccain_names.head()
```

```python
print(df_mccain_names.shape)

## creating df's for replies, retweets, and quotes only. Can be added together later

#creating a df of retweets only
df_mccain_names_rts = df_mccain_names.dropna(axis=0, subset=['retweet_screen_name'])
print(df_mccain_names_rts.shape)

#creating a df of replies only
df_mccain_names_replies = df_mccain_names.dropna(axis=0, subset=['reply_to_screen_name'])
print(df_mccain_names_replies.shape)

#creating a df of quotes only
df_mccain_names_quotes = df_mccain_names.dropna(axis=0, subset=['quoted_screen_name'])
print(df_mccain_names_quotes.shape)
```

Just a note - come back to the above if I need to add more information to my network data

Also - I'm just going to go with retweets first this time. Retweets more often imply someone is agreeing with what is being said, while a quote is often someone disagreeing

```python
df_mccain_names_rts.sample(10)
```

```python
# First creating undirected:
G_mccain_rt_und = nx.from_pandas_edgelist(df_mccain_names_rts, source='screen_name', target='retweet_screen_name', edge_attr=True, create_using=None)

# Second creating directed:
G_mccain_rt_dir = nx.from_pandas_edgelist(df_mccain_names_rts, source='screen_name', target='retweet_screen_name', edge_attr=True, create_using=nx.DiGraph())
```

```python
print(nx.info(G_mccain_rt_und))
print(nx.info(G_mccain_rt_dir))
```

```python
# removing all self re-tweets
G_mccain_rt_und.remove_edges_from(G_mccain_rt_und.selfloop_edges())

```

```python
print(nx.info(G_mccain_rt_und))
print(nx.info(G_mccain_rt_dir))
```

That removed a few thousand - also, I am going to stick with undirected, because the next step undirects it anway

```python
## The biggest component is way bigger than every other one, so we can just focus on that for the graph
G_mccain_rt_und_subgraphs = sorted(nx.connected_component_subgraphs(G_mccain_rt_und.to_undirected()), key=len, reverse=True)
print('Size of largest connected components:', [len(g) for g in G_mccain_rt_und_subgraphs[:20]])
```

```python
G_mccain_rt_dir_subgraphs = sorted(nx.connected_component_subgraphs(G_mccain_rt_dir.to_undirected()), key=len, reverse=True)
print('Size of largest connected components:', [len(g) for g in G_mccain_rt_dir_subgraphs[:20]])
```

```python
G_mccain_rt_und_final = G_mccain_rt_und_subgraphs[0]
```

```python
G_mccain_rt_dir_final = G_mccain_rt_dir_subgraphs[0]
```

```python
print(nx.info(G_mccain_rt_und_final))
```

```python
#Directed graph of strongest connected component for directed graph (USEING THIS FOR FINAL ANALYSIS, FEED INTO S FOR STRONGEST COMPONENT)
Gc_mccain_dir_rt = max(nx.strongly_connected_component_subgraphs(G_mccain_rt_dir), key=len)
print(nx.info(Gc_mccain_dir_rt))
```
