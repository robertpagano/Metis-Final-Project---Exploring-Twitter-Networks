```python
from textblob import TextBlob, Word

import pandas as pd
import numpy as np

import pickle

pd.set_option('display.max_colwidth', 400)  # or 199
pd.set_option('display.max_columns', None)  
```

First, run polarity and subjectivity on all tweets, then, append SCC status so we can look at average polarity inside and outside of the red text

```python
with open('/Users/robertpagano/metis_data/project_5/aoc_nlp_null_class_removed.pickle', 'rb') as f:
    df_aoc_all = pickle.load(f)
```

```python
with open('/Users/robertpagano/metis_data/project_5/tuck_nlp_null_class_removed.pickle', 'rb') as f:
    df_tucker_all = pickle.load(f)
```

```python
with open('/Users/robertpagano/metis_data/project_5/marvel_nlp_null_class_removed.pickle', 'rb') as f:
    df_marvel_all = pickle.load(f)
```

```python
df_tucker_all.head()
```

```python
def choose_text(df):
    # selects the full text version of the tweet for analysis
    df['final_text'] = np.where((df['retweet_text'].isnull() == False), df['retweet_text'], df['main_text'])
    return df
```

```python
df_tucker_all = choose_text(df_tucker_all)
df_tucker_all.head(10)
```

```python
df_aoc_all = choose_text(df_aoc_all)
df_marvel_all = choose_text(df_marvel_all)
```

```python
df_aoc_all['polarity'] = df_aoc_all.final_text.apply(lambda x: TextBlob(x).sentiment.polarity)
df_aoc_all['subjectivity'] = df_aoc_all.final_text.apply(lambda x: TextBlob(x).sentiment.subjectivity)
```

```python
df_tucker_all['polarity'] = df_tucker_all.final_text.apply(lambda x: TextBlob(x).sentiment.polarity)
df_tucker_all['subjectivity'] = df_tucker_all.final_text.apply(lambda x: TextBlob(x).sentiment.subjectivity)
```

```python
df_marvel_all['polarity'] = df_marvel_all.final_text.apply(lambda x: TextBlob(x).sentiment.polarity)
df_marvel_all['subjectivity'] = df_marvel_all.final_text.apply(lambda x: TextBlob(x).sentiment.subjectivity)
```

```python
with open('/Users/robertpagano/metis_data/project_5/sentiment_data/df_aoc_all_sentiment.pickle', 'wb') as to_write:
    pickle.dump(df_aoc_all, to_write)
    
with open('/Users/robertpagano/metis_data/project_5/sentiment_data/df_tucker_all_sentiment.pickle', 'wb') as to_write:
    pickle.dump(df_tucker_all, to_write)
    
with open('/Users/robertpagano/metis_data/project_5/sentiment_data/df_marvel_all_sentiment.pickle', 'wb') as to_write:
    pickle.dump(df_marvel_all, to_write)
    
```
